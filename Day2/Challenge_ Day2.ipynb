```python
import re
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Filling NaN using mode
df['hashtags'].fillna(df['hashtags'].mode()[0],inplace=True)
df['place_coord_boundaries'].fillna(df['place_coord_boundaries'].mode()[0],inplace=True)
df['possibly_sensitive'].fillna(df['possibly_sensitive'].mode()[0],inplace=True)
```

```python
#place column has Nan in all row so we have to drop it   
df = df.drop(columns=['place'])
```

```python
# inspecting the polarity column

df['polarity'].value_counts()
```

```python
# checking if there are null values
text_null_sum = df['original_text'].isnull().sum()
polarity_null_sum = df['polarity'].isnull().sum()

print(f'sum of null values in oringal_text column: {text_null_sum}')
print(f'sum of null values in polarity cloumn: {polarity_null_sum}')
```

```python
# explore if there are another languages
df['lang'].value_counts()
```

```python
# drop all rows having language different from english
df = df.drop(df[df['lang'] != 'en'].index)
```

```python
# creating cleanedTweet dataframe
cleanedTweets = pd.DataFrame(columns=['clean_text','polarity'])
cleanedTweets['clean_text'] = df['original_text']
cleanedTweets['polarity'] = df['polarity']
cleanedTweets.dropna(inplace=True)
```

```python
# replacing hastags, username starting by @ and links
def clean_text(text):
    hashtag_removed = re.sub('(#[A-Za-z]+[A-Za-z0-9-_]+)', '', text)
    removed_links = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text, flags=re.MULTILINE)
    cleaned = re.sub('(@[A-Za-z]+[A-Za-z0-9-_]+)', '', removed_links)
    return cleaned
```

```python
# cleaning and inspecting the whole dataframe
cleanedTweets['clean_text'] = cleanedTweets['clean_text'].apply(clean_text)
cleanedTweets
```

```python
df.polarity = df.polarity.astype(float)
```

```python
# their data types
df.dtypes
```

```python
# A function takes polarity and returns a string either postive, negative or neutral
def text_category(p: float) -> str:
    if float(p) > 0:
        return "positive"
    elif float(p):
        return "neutral"
    else:
        return "negative"
```

```python
# creating score column by applying text_category
cleanedTweets['score'] = cleanedTweets['polarity'].apply(text_category)
```

```python
cleanedTweets['score'].value_counts()
```

```python
# Visualizing data by plotting bar graph
score_count =  cleanedTweets['score'].value_counts()
score_count.plot(kind="bar", y="score", title="score", figsize=[8, 8]);
```

```python
score_count.plot(kind="pie", y="score", title="score", figsize=[8, 8]);
```

```python
# This function identifies which row is neutral from score column
def remove_neutral(value):
    return value != "neutral"
```

```python
# using the above function we will remove all rows that have neutral score
# then create a score_map. if score is "positive" it has value 1 else 0
cleanedTweets = cleanedTweets[cleanedTweets['score'].map(remove_neutral)]
cleanedTweets['score_map'] = cleanedTweets["score"].map( lambda score: 1 if score == "positive" else 0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.ensemble import RandomForestClassifier
from joblib import dump, load 
from scipy.sparse import save_npz, load_npz 
from sklearn.decomposition import NMF, LatentDirichletAllocation
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
```

```python
# Splitting data by giving the input to X and the output to y
(X, y) = cleanedTweets['clean_text'], cleanedTweets['score_map']
```

```python
# first we create CountVectorizer and then fit the data 
trigram_vectorizer = CountVectorizer(ngram_range=(1, 3))
trigram_vectorizer.fit(X.values)
```

```python
# creating a tri_gram Vector representation for X
X_trigram = trigram_vectorizer.transform(X.values)
```

```python
# creating a trigram tfidf transformer
trigram_tf_idf_transformer = TfidfTransformer()
trigram_tf_idf_transformer.fit(X_trigram)
```

```python
# creating a tfidf representation of x
X_train_tf_idf = trigram_tf_idf_transformer.transform(X_trigram)
```

```python
# The function takes input as X, and ouput as Y, then split the input to train and valid 
# then use randomforestclassifier to fit the input

def train_and_show_scores(X, y, title: str) -> None:
    X_train, X_valid, y_train, y_valid = train_test_split(
        X, y, train_size=0.75, stratify=y
    )

    clf = RandomForestClassifier(max_depth=2, random_state=0)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    valid_score = clf.score(X_valid, y_valid)
    print(f'{title}\nTrain score: {round(train_score, 2)} ; Validation score: {round(valid_score, 2)}\n')
```

```python
# train using inputs vectorized as trigram

train_and_show_scores(X_trigram, y, title="trigram")
```

```python
# train using inputs vectorized as trigram_tfidf
train_and_show_scores(X_train_tf_idf, y, title="trigram_tf_idf")
```
